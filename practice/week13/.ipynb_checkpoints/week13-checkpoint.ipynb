```python
import matplotlib.pyplot as plt
import csv
```

## plot(), bar(), barh()

```python
xdata = [1,2,3,4]
ydata = [10,5,20,30]

plt.plot(xdata, ydata)
plt.show()

plt.bar(xdata, ydata)
plt.show()

plt.barh(xdata, ydata)
plt.show()
```

## 실습: 내 생일 서울의 최고 기온 시각화

```python
f = open('seoul.csv')
data =csv.reader(f)

next(data)
result = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '12' and row[0].split('-')[2] =='15':
            result.append(float(row[-1]))

plt.plot(result)
plt.show()
```

## 실습: 기온 데이터를 히스토그램으로 표현하기

```python
f = open('seoul.csv')
data =csv.reader(f)
next(data)
aug = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month =='08':
            aug.append(float(row[-1]))
            
# bins? 빈도를 표현한 것이므로 실수일 경우 전부 다른 값이라 모두 1일 수 있다.
# min, max값을 100단위 쪼개어 처리            
plt.hist(aug, bins=100, color='r')
plt.show()
```

## 실습: 기온 데이터를 상자그림으로 표현하기

```python
f = open('seoul.csv')
data =csv.reader(f)
next(data)
month = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[-1] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))

plt.boxplot(month)
plt.show()
```

## 기본 그래프 그리기
- pie(), scatter()

```python
data = [10, 20, 30]
plt.pie(data)
plt.show()

xdata = [1,2,3,4]
ydata = [10,40,20,30]
plt.scatter(xdata, ydata)
plt.show()
```

## 그래프 그리기 옵션 - 1
- plot() 옵션
    - 색상:r, g, b
    - 모양:., ^ 
        - r. , r^
    - 레이블: label
        
- 공통 옵션
    - 제목 표시: title()
    - 범례 표시: legend()

- pie() 옵션
    - 레이블: labels
    - 색상: colors
    - 비율: autopct
    - 돌출: explode
    

```python
data1 = [1,2,3,4]
data2 = [4,3,1,1]

plt.title('plot graph')
plt.plot(data1,'r.',label='circle')
plt.plot(data2,'g^',label='triangle')
plt.legend() # label 생성
plt.show()
```


## 그래프 그리기 옵션 - 2
- pie() 옵션
    - 레이블: labels
    - 색상: colors
    - 비율: autopct
    - 돌출: explode

```python
data = [10,30,20,45]
label = ['A', 'B', 'C', 'D']
color = ['hotpink', 'darkmagenta', 'pink', 'deeppink']

plt.pie(data, labels=label, autopct='%.6f%%', colors=color, explode =[0,0,0.1,0])
plt.legend()
plt.show()
```

## 그래프 그리기 옵션 - 2
- scatter() 옵션
    - 컬러바 표시: colorbar()
    - 크기: s
    - 색상 범위: c
    - 색상 맵: cmap

```python
xdata = [1,2,3,4]
ydata = [10,40,20,30]
sizes = [10,50,100,200]

plt.scatter(xdata, ydata, s=sizes, c=range(4), cmap='jet')
plt.colorbar()
plt.show()
```

## 실습: 항아리 모양 그래프 그리기
- [ www.mois.go.kr ] - [정책자료] - [통계] - [주민등록 인구통계]

```python
f = open('gender.csv')
data = csv.reader(f)

m = [] # 1살 ~ 100살 남자 인구
f = [] # 1살 ~ 100살 여자 인구

for row in data:
    if '신도림' in row[0]:
        for i in range(0, 101):
            m.append(-int(row[i+3]))
            f.append(int(row[-(i+1)]))
f.reverse()


            
#         for i in row[3:104]: # 1살~ 100살
#             m.append(int(i))
#         for i in row[106:]: # 1살 ~ 100살
#             f.append(int(i)) 

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title('신도림 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, lable='남성')
plt.barh(range(101), f, lable='여성')

plt.legend()
plt.show()

            
```
