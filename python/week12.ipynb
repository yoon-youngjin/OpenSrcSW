실습 1

```python
import csv
f = open('seoul.csv')
data = csv.reader(f) # list형태로 return
next(data)
label = [''] * 3
mx = [0] * 3
for row in data:
    for i in range(3, 5):
        row[i] = float(row[i])
    
    if (row[4] - row[3]) > mx[0]:
        mx[0] = (row[4] - row[3])
        label[0] = row[0]
        
    if row[3] < mx[2]:
        mx[2] = row[3]
        label[1] = row[0]
    
    if row[4] > mx[1]:
        mx[1] = row[4]
        label[2] = row[0]
                       
f.close()
        
print('일교차가 가장 큰 날')
print(label[0])

print('일교차가 가장 큰 날')
print(label[1])

print('일교차가 가장 큰 날')
print(label[2])
```

```python
f = open('subwayfee.csv')
data = csv.reader(f) # list형태로 return
next(data)

line={}
for row in data:
    station={}
    station[row[3] + ' ' + row[1]] = int(row[4].replace(',',''))+int(row[5].replace(',',''))
    if row[1] in line:
        line[row[1]].update(station)
        
    else:
        line[row[1]] = station
        
for station in line:
    print('이용객이 가장 많은 역 순위')
    sorted_list = sorted(line[station].items(), key=lambda x:x[1], reverse=True)
    for i in range(3):
        print(str(i+1)+'. '+sorted_list[i][0]+'('+str(sorted_list[i][1])+'명)')
    print('이용객이 가장 적은 역 순위')
    for j in range(3):
        print(str(j+1)+'. '+sorted_list[-(j+1)][0]+'('+str(sorted_list[-(j+1)][1])+'명)')
    print()    
```
